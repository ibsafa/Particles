```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# DEFINE MY 1/ALPHA FUNCTION #

```python
def get_alpha(B_i, mu):
    A = -8.*np.pi/B_i
    B = np.log(mu)
    return B/A
```

# Part B#

```python
mu = np.logspace(1, 19, 200)
alpha1 = []
alpha2 = []
alpha3 = []
B1 = 41./10.
B2 = -19./6.
B3 = -7.
for x in range(len(mu)):
    alpha1.append(get_alpha(B1, mu[x]) -(-0.714856 - 0.0169))
    alpha2.append(get_alpha(B2, mu[x]) - (0.5521 - 0.03383))
    alpha3.append(get_alpha(B3,mu[x]) - (1.22049 - 0.1185))
```

```python
plt.figure(figsize=[8,8])
plt.plot(mu,alpha1, label='$ 1/\\alpha_{1}$')
plt.plot(mu,alpha2, label='$ 1/\\alpha_{2}$')
plt.plot(mu,alpha3, label='$ 1/\\alpha_{3}$')
plt.axvline(80, linestyle = '--', color='black', label='$ 80 GeV$', alpha = 0.5)
plt.semilogx()

plt.xlabel('$Energy (GeV) $', fontsize = 14)
plt.title('$ 1/ \\alpha\ vs\ Energy $', fontsize = 18)
plt.legend()
plt.show()
```

```python
B1 = 33./5.
B2 = 1.
B3 = -3.
```

```python
mu = np.logspace(3, 19, 200)
alpha1 = []
alpha2 = []
alpha3 = []
for x in range(len(mu)):
    alpha1.append(get_alpha(B1, mu[x]) - (-1.81401 + 0.395))
    if(np.log10(mu[x]) > 7.):
        alpha2.append(get_alpha(B2, mu[x]) - (-0.27485 - 0.352))
    alpha3.append(get_alpha(B3,mu[x]) - (0.82455 - 0.82197))
```

```python
plt.figure(figsize=[8,8])
plt.plot(mu,1./np.asarray(alpha1),label='$ \\alpha_{1}$')
plt.plot(mu[50:],1./np.asarray(alpha2),label='$ \\alpha_{2}$')
plt.plot(mu,1./np.asarray(alpha3),label='$ \\alpha_{3}$')
plt.semilogx()
plt.legend()
plt.xlim(1e3, 1e19)
plt.ylim(-5,5)
plt.xlabel('$ \ Energy (GeV) $',fontsize=16)
plt.ylabel('$ \\alpha_{i} $',fontsize=16)
plt.title('$ \\alpha\ vs\ Energy $', fontsize = 18)

plt.show()
```
